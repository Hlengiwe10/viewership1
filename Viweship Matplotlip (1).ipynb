```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel(
    '/Workspace/Users/kamagaduzela10@gmail.com/Viewership Analysis .xlsx'
)

# Pivot the data for grouped bar chart
pivot_df = df.pivot_table(
    index='Platform',
    columns='PlayEventType',
    values='TotalTimeWatched',
    aggfunc='sum'
)

pivot_df.plot(
    kind='bar',
    figsize=(10, 6)
)
plt.xlabel('Platform')
plt.ylabel('TotalTimeWatched')
plt.title('Grouped Bar Chart of TotalTimeWatched by Platform and PlayEventType')
plt.legend(title='PlayEventType')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel(
    '/Workspace/Users/kamagaduzela10@gmail.com/Viewership Analysis .xlsx'
)

# Load Excel file
#file_path = '/Workspace/Users/kamagaduzela10@gmail.com/Viewership Analysis.xlsx'
#df = pd.read_excel(file_path)

# Group by Platform and sum TotalTimeWatched
df_grouped = df.groupby('Platform', as_index=False)['TotalTimeWatched'].sum()

# Sort for clarity
df_grouped = df_grouped.sort_values('TotalTimeWatched', ascending=False)

# Create pie chart
plt.figure(figsize=(8, 6))
plt.pie(
    df_grouped['TotalTimeWatched'],
    labels=df_grouped['Platform'],
    autopct='%1.1f%%',
    startangle=140
)

plt.title('Total Time Watched by Platform')
plt.tight_layout()
plt.show()
```
